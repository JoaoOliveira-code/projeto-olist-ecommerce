```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DATA_PATH = "../data/"

customers = pd.read_csv(DATA_PATH + "olist_customers_dataset.csv")
orders = pd.read_csv(DATA_PATH + "olist_orders_dataset.csv")
products = pd.read_csv(DATA_PATH + "olist_products_dataset.csv") 

print("Clientes (Customers):")
display(customers.head())

print("\nPedidos (Orders):")
display(orders.head())

print("\nProdutos (Products):")
display(products.head())
```

```python
orders_df = orders.copy()

date_columns = ['order_purchase_timestamp', 'order_approved_at', 
                'order_delivered_carrier_date', 'order_delivered_customer_date', 
                'order_estimated_delivery_date']

for col in date_columns:
    orders_df[col] = pd.to_datetime(orders_df[col])

print("Tipos de dados após a conversão:")
orders_df.info()
```

```python
print(f"Total de pedidos sem data de entrega: {orders_df['order_delivered_customer_date'].isnull().sum()}")


pedidos_sem_entrega = orders_df[orders_df['order_delivered_customer_date'].isnull()]

print("\nStatus dos pedidos sem data de entrega:")
display(pedidos_sem_entrega['order_status'].value_counts())
```

```python
df_completo = pd.merge(orders_df, customers, on='customer_id')
display(df_completo.head())
```

```python
# Definir um estado visual mais agradavel para os gráficos
sns.set_style("whitegrid")

#Criar a figura que vai abrigar o gráfico
plt.figure(figsize=(15, 6))

#Contar os pedidos por estado e criar um gráfico de barras
sns.countplot(x='customer_state', data=df_completo, order=df_completo['customer_state'].value_counts().index)

plt.title('Distribuição de Pedidos por Estado', fontsize=16)

#Melhorar os nomes dos eixos
plt.xlabel('Estado', fontsize=12)
plt.ylabel('Número de Pedidos', fontsize=12)

#Exibir o gráfico
plt.show()
```

```python
#Garantir que a coluna de data de compra está no formato correto 
df_completo['order_purchase_timestamp'] = pd.to_datetime(df_completo['order_purchase_timestamp'])

df_completo['purchase_month'] = df_completo['order_purchase_timestamp'].dt.to_period('M')

dias_semana_map = {
    'Monday': 'Segunda-feira', 'Tuesday': 'Terça-feira', 'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira', 'Friday': 'Sexta-feira', 'Saturday': 'Sábado', 'Sunday': 'Domingo'
}
df_completo['purchase_day_of_week'] = df_completo['order_purchase_timestamp'].dt.day_name().map(dias_semana_map)

display(df_completo[['order_purchase_timestamp', 'purchase_month', 'purchase_day_of_week']].head())
```

```python
monthly_orders = df_completo['purchase_month'].value_counts().sort_index()

# O .sort_index() é crucial para garantir que os meses fiquem em ordem cronológica

plt.figure(figsize=(15, 6))
monthly_orders.plot(kind='line', marker='o')

plt.title('Evolução do Número de Pedidos por Mês', fontsize=16)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Número de Pedidos', fontsize=12)
plt.grid(True) # Adiciona uma grade para facilitar a leitura
plt.show()                 
```

```python
#Definir a ordem correta dos dias da semana para o gráfico
ordem_dias = ['Segunda-feira', 'Terça-feira','Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo']

plt.figure(figsize=(12, 6))
sns.countplot(x='purchase_day_of_week', data=df_completo, order=ordem_dias, palette='viridis')

plt.title('Número de Pedidos por Dia da Semana', fontsize=16)
plt.xlabel('Dia da Semana', fontsize=12)
plt.ylabel('Número de Pedidos', fontsize=12)
plt.xticks(rotation=45) # Rotaciona os nomes dos dias para não sobrepor
plt.show()
```

```python
# Carregar os datasets que faltam
payments = pd.read_csv(DATA_PATH + "olist_order_payments_dataset.csv")
order_items = pd.read_csv(DATA_PATH + "olist_order_items_dataset.csv")
products_info = pd.read_csv(DATA_PATH + "olist_products_dataset.csv")
translation = pd.read_csv(DATA_PATH + "product_category_name_translation.csv")

# Juntar as informações de pagamento ao nosso dataframe principal
df_completo = pd.merge(df_completo, payments, on='order_id')

# Juntar as informações dos itens do pedido
df_completo = pd.merge(df_completo, order_items, on='order_id')

# Juntar as informações dos produtos
df_completo = pd.merge(df_completo, products_info, on='product_id')

# Juntar a tradução das categorias
df_completo = pd.merge(df_completo, translation, on='product_category_name')

# Vamos dar uma espiada no resultado final. Agora temos uma tabela gigante com tudo!
print("Colunas do DataFrame final:")
print(df_completo.columns)
display(df_completo.head())
```

```python
plt.figure(figsize=(12, 6))

#Contar os tipos de pagamento e plotar o gráfico
sns.countplot(y='payment_type', data=df_completo, order=df_completo['payment_type'].value_counts().index, palette='magma')

plt.title('Formas de Pagamento Mais Utilizadas', fontsize=16)
plt.xlabel('Número de Transações', fontsize=12)
plt.ylabel('Forma de Pagamento', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 10))

#Contar as categorias e plotar o gráfico. ('product_category_name_english')
sns.countplot(y='product_category_name_english', data=df_completo,
             order=df_completo['product_category_name_english'].value_counts().nlargest(15).index,
             palette='rocket')

plt.title('Top 15 Categorias de Produtos Mais Vendidas', fontsize=16)
plt.xlabel('Número de Vendas', fontsize=12)
plt.ylabel('Categoria do Produto', fontsize=12)
plt.show()
```
